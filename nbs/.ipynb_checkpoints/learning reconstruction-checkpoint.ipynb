```python
# def compute_rotation_matrix_from_vectors(vec1, vec2):
#     """
#     Compute the rotation matrix that aligns vec1 to vec2.

#     Parameters:
#         vec1 (ndarray): Source vector.
#         vec2 (ndarray): Destination vector.

#     Returns:
#         rotation_matrix (ndarray): The rotation matrix.
#     """
#     a = vec1 / np.linalg.norm(vec1)
#     b = vec2 / np.linalg.norm(vec2)
#     cross = np.cross(a, b)
#     dot = np.dot(a, b)
#     if np.isclose(dot, -1.0):
#         # Vectors are opposite
#         ortho = np.array([1, 0, 0]) if not np.allclose(a, [1, 0, 0]) else np.array([0, 1, 0])
#         rotation_matrix = -np.eye(3) + 2 * np.outer(ortho, ortho)
#     elif np.isclose(dot, 1.0):
#         # Vectors are the same
#         rotation_matrix = np.eye(3)
#     else:
#         skew = np.array([[0, -cross[2], cross[1]],
#                          [cross[2], 0, -cross[0]],
#                          [-cross[1], cross[0], 0]])
#         rotation_matrix = np.eye(3) + skew + np.matmul(skew, skew) * ((1 - dot) / (np.linalg.norm(cross) ** 2))
#     return rotation_matrix


# def align_floor_to_xz(pcd, plane_model):
#     # Initial detected plane normal
#     plane_normal = np.array(plane_model[:3])
#     plane_normal = plane_normal / np.linalg.norm(plane_normal)
#     desired_normal = np.array([1.0, 0.0, 1.0])  # Align to YZ plane

#     # Compute the rotation matrix to align plane normal to desired normal
#     rotation_matrix = compute_rotation_matrix_from_vectors(plane_normal, desired_normal)

#     # Apply the rotation to the point cloud
#     rotated_pcd = copy.deepcopy(pcd).rotate(rotation_matrix, center=(0, 0, 0))

#     # Re-detect the plane after rotation
#     plane_model_aligned, inliers = detect_floor_plane(rotated_pcd)
#     if plane_model_aligned is None:
#         raise ValueError("No floor plane detected after rotation.")
#     [a, b, c, d] = plane_model_aligned
#     print(f"Plane equation after rotation: {a:.1f}x + {b:.1f}y + {c:.1f}z + {d:.1f} = 0")

#     # Force alignment to the desired plane (YZ)
#     residual_normal = np.array(plane_model_aligned[:3])
#     residual_normal = residual_normal / np.linalg.norm(residual_normal)
#     refinement_matrix = compute_rotation_matrix_from_vectors(residual_normal, desired_normal)
#     rotated_pcd.rotate(refinement_matrix, center=(0, 0, 0))

#     # Alignment to origin
#     plane_model_final, inliers = detect_floor_plane(rotated_pcd)
#     [a, b, c, d] = plane_model_final
#     print(f"Final plane equation: {a:.1f}x + {b:.1f}y + {c:.1f}z + {d:.1f} = 0")

#     # Translate the plane to pass through the origin (d=0)
#     translation = np.array([0, -d / a if a != 0 else 0, 0])  # Align along X-axis
#     rotated_pcd.translate(translation)

#     # Center the entire point cloud's centroid at the origin
#     centroid = np.mean(np.asarray(rotated_pcd.points), axis=0)
#     transformed_pcd = rotated_pcd.translate(-centroid)

#     # Verify final alignment
#     final_plane_model, _ = detect_floor_plane(transformed_pcd)
#     [a, b, c, d] = final_plane_model
#     print(f"Verified final plane equation: {a:.2f}x + {b:.2f}y + {c:.2f}z + {d:.2f} = 0")
#     final_centroid = np.mean(np.asarray(_.points), axis=0)
#     print(f"Final centroid: {final_centroid}")
#     if np.allclose(a, 0, atol=1e-2) and np.allclose(b, 1, atol=1e-2) and np.allclose(c, 0, atol=1e-2):
#         return transformed_pcd
#     elif final_plane_model is None: 
#         return transformed_pcd
#     else:
#         align_floor_to_xz(transformed_pcd, final_plane_model)


# # import numpy as np
# # import open3d as o3d
# # import copy

# # def align_floor_to_xz(pcd, plane_model):
# #     """
# #     Align the detected floor plane to the XZ plane (Y=0) and center it at the origin.

# #     Parameters:
# #         pcd (o3d.geometry.PointCloud): The input point cloud.
# #         plane_model (list): The coefficients of the plane equation [a, b, c, d].

# #     Returns:
# #         transformed_pcd (o3d.geometry.PointCloud): The transformed point cloud.
# #     """
# #     # Step 1: Compute the plane normal and desired normal
# #     plane_normal = np.array(plane_model[:3])
# #     plane_normal = plane_normal / np.linalg.norm(plane_normal)  # Normalize the plane normal
# #     desired_normal = np.array([0.0, 1.0, 0.0])  # Desired normal for alignment with the XZ plane

# #     # Step 2: Compute the rotation matrix to align the plane normal to the desired normal
# #     rotation_matrix = compute_rotation_matrix_from_vectors(plane_normal, desired_normal)

# #     # Step 3: Rotate the point cloud
# #     rotated_pcd = copy.deepcopy(pcd)
# #     rotated_pcd.rotate(rotation_matrix, center=(0, 0, 0))

# #     # Step 4: Detect the plane again after rotation
# #     plane_model_aligned, inliers = detect_floor_plane(rotated_pcd)
# #     if plane_model_aligned is None:
# #         raise ValueError("Failed to detect the floor plane after rotation.")

# #     # Step 5: Select floor points and compute their centroid
# #     floor_points = rotated_pcd.select_by_index(inliers)
# #     floor_centroid = np.mean(np.asarray(floor_points.points), axis=0)

# #     # Step 6: Translate the entire point cloud to center the floor plane at the origin
# #     translation_vector = -floor_centroid  # Move the floor centroid to the origin
# #     transformed_pcd = copy.deepcopy(rotated_pcd)
# #     transformed_pcd.translate(translation_vector)

# #     # Step 7: Validate final alignment and centroid
# #     final_plane_model, final_inliers = detect_floor_plane(transformed_pcd)
# #     if final_plane_model is None:
# #         raise ValueError("Failed to detect the floor plane after final translation.")

# #     # Verify final plane alignment
# #     [a, b, c, d] = np.round(final_plane_model, 2)
# #     print(f"Final plane equation: {a:.2f}x + {b:.2f}y + {c:.2f}z + {d:.2f} = 0")

# #     # Compute final centroid of all points (should be at origin)
# #     final_centroid = np.mean(np.asarray(transformed_pcd.points), axis=0)
# #     print(f"Final centroid: {final_centroid}")

# #     # Return the transformed point cloud
# #     return transformed_pcd

import numpy as np
import open3d as o3d
import copy

def align_plane_to_y_axis(pcd, plane_model):
    """
    Align the detected plane to lie along y=0.

    Parameters:
        pcd (o3d.geometry.PointCloud): Input point cloud.
        plane_model (list): Plane equation coefficients [a, b, c, d].

    Returns:
        transformed_pcd (o3d.geometry.PointCloud): The transformed point cloud.
    """
    # Step 1: Normalize the plane normal
    plane_normal = np.array(plane_model[:3])
    plane_normal = plane_normal / np.linalg.norm(plane_normal)  # Normalize the normal vector
    desired_normal = np.array([0.0, 1.0, 0.0])  # Normal for the plane lying on y=0

    # Step 2: Compute rotation matrix
    rotation_matrix = compute_rotation_matrix_from_vectors(plane_normal, desired_normal)

    # Step 3: Rotate the point cloud
    rotated_pcd = copy.deepcopy(pcd)
    rotated_pcd.rotate(rotation_matrix, center=(0, 0, 0))

    # Step 4: Re-detect the plane and compute translation
    plane_model_aligned, inliers = detect_floor_plane(rotated_pcd)
    if plane_model_aligned is None:
        raise ValueError("Failed to detect the plane after rotation.")

    # Calculate the translation vector to align the plane along y=0
    d_aligned = plane_model_aligned[3]  # Use the updated d from the plane equation
    translation_vector = np.array([0, -d_aligned, 0])  # Translate along Y-axis

    # Apply translation
    transformed_pcd = copy.deepcopy(rotated_pcd)
    transformed_pcd.translate(translation_vector)

    # Step 5: Verify the final alignment
    final_plane_model, _ = detect_floor_plane(transformed_pcd)
    if final_plane_model is None:
        raise ValueError("Failed to detect the plane after translation.")

    [a, b, c, d] = np.round(final_plane_model, 2)
    print(f"Final plane equation: {a:.2f}x + {b:.2f}y + {c:.2f}z + {d:.2f} = 0")

    # Compute the final centroid
    final_centroid = np.mean(np.asarray(transformed_pcd.points), axis=0)
    print(f"Final centroid: {final_centroid}")

    return transformed_pcd
```

```python
import os
import copy
import numpy as np
import open3d as o3d

np.random.seed(2024)

def load_point_cloud(file_path):
    """
    Load a point cloud from a file.

    Parameters:
        file_path (str): The path to the point cloud file.

    Returns:
        pcd (o3d.geometry.PointCloud): The loaded point cloud.
    """
    # Check if file exists
    if not os.path.isfile(file_path):
        print(f"File not found: {file_path}")
        return None

    # Load the point cloud
    pcd = o3d.io.read_point_cloud(file_path)
    print("Point cloud loaded.")
    print(f"Number of points: {len(pcd.points)}")

    # Check if point cloud is empty
    if len(pcd.points) == 0:
        print("Point cloud is empty. Please check the input file.")
        return None
    
    return pcd

def downsample_point_cloud(pcd, voxel_size=0.007):
    """
    Downsample the point cloud using a voxel grid filter.

    Parameters:
        pcd (o3d.geometry.PointCloud): The input point cloud.
        voxel_size (float): The size of the voxel grid.

    Returns:
        down_pcd (o3d.geometry.PointCloud): The downsampled point cloud.
    """
    down_pcd = pcd.voxel_down_sample(voxel_size=voxel_size)
    return down_pcd

def estimate_normals(pcd, radius=0.05, max_nn=200):
    """
    Estimate normals for the point cloud.

    Parameters:
        pcd (o3d.geometry.PointCloud): The input point cloud.
        radius (float): The search radius for normal estimation.
        max_nn (int): Maximum number of nearest neighbors to use.

    Returns:
        None (The point cloud's normals are updated in place.)
    """
    if len(pcd.points) == 0:
        print("Point cloud is empty. Cannot estimate normals.")
        return
    try:
        pcd.estimate_normals(search_param=o3d.geometry.KDTreeSearchParamHybrid(
            radius=radius, max_nn=max_nn))
        # Check if normals are computed
        if not pcd.has_normals():
            print("Normals were not estimated.")
            return
        # Orient normals consistently
        pcd.normalize_normals()
        pcd.orient_normals_towards_camera_location(camera_location=np.array([10, 10., 10.]))
        pcd.orient_normals_consistent_tangent_plane(k=25)
    except Exception as e:
        print(f"Error estimating normals: {e}")

def remove_statistical_outliers(pcd, nb_neighbors=10, std_ratio=0.01):
    """
    Remove statistical outliers from the point cloud.

    Parameters:
        pcd (o3d.geometry.PointCloud): The input point cloud.
        nb_neighbors (int): Number of neighbors to analyze for each point.
        std_ratio (float): Standard deviation ratio.

    Returns:
        inlier_cloud (o3d.geometry.PointCloud): The filtered point cloud.
    """
    cl, ind = pcd.remove_statistical_outlier(nb_neighbors=nb_neighbors,
                                             std_ratio=std_ratio)
    inlier_cloud = pcd.select_by_index(ind)
    return inlier_cloud

def detect_floor_plane(pcd, distance_threshold=0.01, ransac_n=5, num_iterations=1000):
    plane_model, inliers = pcd.segment_plane(distance_threshold=distance_threshold,
                                             ransac_n=ransac_n,
                                             num_iterations=num_iterations)
    [a, b, c, d] = np.round(plane_model, 2)
    print(f"Plane equation: {a:.2f}x + {b:.2f}y + {c:.2f}z + {d:.2f} = 0")
    if len(inliers) == 0:
        print("No plane detected.")
        return None, None
    return plane_model, inliers

def detect_floor_plane_with_normal(pcd, expected_normal, angle_threshold=np.pi / 4):
    max_iterations = 3
    for _ in range(max_iterations):
        plane_model, inliers = pcd.segment_plane(distance_threshold=0.01,
                                                 ransac_n=3,
                                                 num_iterations=1000)
        if len(inliers) == 0:
            break
        normal = np.array(plane_model[:3])
        normal /= np.linalg.norm(normal)
        angle = np.arccos(np.clip(np.dot(normal, expected_normal), -1.0, 1.0))
        if angle < angle_threshold:
            return plane_model, inliers
        # Remove the detected plane and continue
        pcd = pcd.select_by_index(inliers, invert=True)
    return None, None

# Compute rotation matrix
def compute_rotation_matrix_from_vectors(v1, v2):
    """Compute the rotation matrix that aligns v1 to v2."""
    a, b = (v1 / np.linalg.norm(v1)).reshape(3), (v2 / np.linalg.norm(v2)).reshape(3)
    v = np.cross(a, b)
    c = np.dot(a, b)
    s = np.linalg.norm(v)
    k_mat = np.array([[0, -v[2], v[1]], [v[2], 0, -v[0]], [-v[1], v[0], 0]])
    rotation_matrix = np.eye(3) + k_mat + k_mat @ k_mat * ((1 - c) / (s ** 2 + 1e-10))
    return rotation_matrix

# def compute_rotation_matrix_from_vectors(vec1, vec2):
#     """
#     Compute the rotation matrix that aligns vec1 to vec2.

#     Parameters:
#         vec1 (ndarray): Source vector.
#         vec2 (ndarray): Destination vector.

#     Returns:
#         rotation_matrix (ndarray): The rotation matrix.
#     """
#     a = vec1 / np.linalg.norm(vec1)
#     b = vec2 / np.linalg.norm(vec2)
#     cross = np.cross(a, b)
#     dot = np.dot(a, b)
#     if np.isclose(dot, -1.0):
#         # Vectors are opposite
#         ortho = np.array([1, 0, 0]) if not np.allclose(a, [1, 0, 0]) else np.array([0, 1, 0])
#         rotation_matrix = -np.eye(3) + 2 * np.outer(ortho, ortho)
#     elif np.isclose(dot, 1.0):
#         # Vectors are the same
#         rotation_matrix = np.eye(3)
#     else:
#         skew = np.array([[0, -cross[2], cross[1]],
#                          [cross[2], 0, -cross[0]],
#                          [-cross[1], cross[0], 0]])
#         rotation_matrix = np.eye(3) + skew + np.matmul(skew, skew) * ((1 - dot) / (np.linalg.norm(cross) ** 2))
#     return rotation_matrix

# def align_floor_to_xz(pcd, plane_model):
#     """
#     Align the detected floor plane to the YZ plane (x=0) and center it on Y=0.

#     Parameters:
#         pcd (o3d.geometry.PointCloud): The input point cloud.
#         plane_model (list): The coefficients of the plane equation [a, b, c, d].

#     Returns:
#         transformed_pcd (o3d.geometry.PointCloud): The transformed point cloud.
#     """
#     # Compute the normal of the input plane
#     plane_normal = np.array(plane_model[:3])
#     plane_normal = plane_normal / np.linalg.norm(plane_normal)
#     desired_normal = np.array([0.0, -1.0, 0.0])  # X-axis for alignment to YZ plane

#     rotation_matrix = compute_rotation_matrix_from_vectors(plane_normal, desired_normal)
#     rotated_pcd = copy.deepcopy(pcd)
#     rotated_pcd.rotate(rotation_matrix, center=(0, 0, 0))

#     plane_model_aligned, inliers = detect_floor_plane(rotated_pcd)
#     floor_points = rotated_pcd.select_by_index(inliers)
#     centroid = np.mean(np.asarray(floor_points.points), axis=0)

#     # Translate to center floor on Y=0 (keeping centroid's Y-coordinate as 0)
#     transformed_pcd = copy.deepcopy(rotated_pcd)
#     transformed_pcd.points = o3d.utility.Vector3dVector(np.asarray(transformed_pcd.points) - centroid)

#     return transformed_pcd

def align_floor_to_xz(pcd, plane_model):
    # Initial detected plane normal
    plane_normal = np.array(plane_model[:3])
    plane_normal = plane_normal / np.linalg.norm(plane_normal)
    desired_normal = np.array([0.0, 1.0, 0.0])  # Align to YZ plane

    # Compute the rotation matrix to align plane normal to desired normal
    rotation_matrix = compute_rotation_matrix_from_vectors(plane_normal, desired_normal)

    # Apply the rotation to the point cloud
    rotated_pcd = copy.deepcopy(pcd).rotate(rotation_matrix, center=(0, 0, 0))

    # Re-detect the plane after rotation
    plane_model_aligned, inliers = detect_floor_plane(rotated_pcd)
    if plane_model_aligned is None:
        raise ValueError("No floor plane detected after rotation.")
    [a, b, c, d] = plane_model_aligned
    print(f"Plane equation after rotation: {a:.1f}x + {b:.1f}y + {c:.1f}z + {d:.1f} = 0")

    # Force alignment to the desired plane (YZ)
    residual_normal = np.array(plane_model_aligned[:3])
    residual_normal = residual_normal / np.linalg.norm(residual_normal)
    refinement_matrix = compute_rotation_matrix_from_vectors(residual_normal, desired_normal)
    rotated_pcd.rotate(refinement_matrix, center=(0, 0, 0))

    # Alignment to origin
    plane_model_final, inliers = detect_floor_plane(rotated_pcd)
    [a, b, c, d] = plane_model_final
    print(f"Final plane equation: {a:.1f}x + {b:.1f}y + {c:.1f}z + {d:.1f} = 0")

    # Translate the plane to pass through the origin (d=0)
    translation = np.array([0, -d / a if a != 0 else 0, 0])  # Align along X-axis
    rotated_pcd.translate(translation)

    # Center the entire point cloud's centroid at the origin
    centroid = np.mean(np.asarray(rotated_pcd.points), axis=0)
    transformed_pcd = rotated_pcd.translate(-centroid)

    # Verify final alignment
    final_plane_model, _ = detect_floor_plane(transformed_pcd)
    [a, b, c, d] = final_plane_model
    print(f"Verified final plane equation: {a:.2f}x + {b:.2f}y + {c:.2f}z + {d:.2f} = 0")
    final_centroid = np.mean(np.asarray(transformed_pcd.select_by_index(inliers).points), axis=0)
    print(f"Final centroid: {final_centroid}")
    if np.allclose(a, 0, atol=1e-2) and np.allclose(b, 1, atol=1e-2) and np.allclose(c, 0, atol=1e-2):
        return transformed_pcd
    elif final_plane_model is None: 
        return transformed_pcd
    else:
        align_floor_to_xz(transformed_pcd, final_plane_model)

def surface_reconstruction(pcd, method='poisson', **kwargs):
    """
    Perform surface reconstruction to create a smooth surface.

    Parameters:
        pcd (o3d.geometry.PointCloud): The input point cloud.
        method (str): Reconstruction method ('poisson' or 'bpa').

    Returns:
        mesh (o3d.geometry.TriangleMesh): The reconstructed mesh.
    """
    estimate_normals(pcd)
    if method == 'poisson':
        print(f"Starting Poisson surface reconstruction with depth {kwargs['depth']}")
        mesh, densities = o3d.geometry.TriangleMesh.create_from_point_cloud_poisson(
            pcd, depth=kwargs["depth"], linear_fit=True)
        print("Poisson reconstruction completed.")

        # Remove low-density vertices
        densities = np.asarray(densities)
        density_threshold = np.percentile(densities, 2)
        vertices_to_remove = densities < density_threshold
        mesh.remove_vertices_by_mask(vertices_to_remove)
    elif method == 'bpa':
        print("Starting BPA surface reconstruction")
        distances = pcd.compute_nearest_neighbor_distance()
        avg_dist = np.mean(distances)
        radii = [avg_dist * factor for factor in [1, 2, 4]]
        mesh = o3d.geometry.TriangleMesh.create_from_point_cloud_ball_pivoting(
            pcd, o3d.utility.DoubleVector(radii))
        print("BPA reconstruction completed.")
    elif method == 'alpha_shapes':
        print(f"Starting Alpha shapes reconstruction with alpha {kwargs['alpha']}")
        tetra_mesh, pt_map = o3d.geometry.TetraMesh.create_from_point_cloud(pcd)
        mesh = o3d.geometry.TriangleMesh.create_from_point_cloud_alpha_shape(pcd, kwargs["alpha"]).fill_holes()
    else:
        raise ValueError("Invalid reconstruction method. Choose 'poisson' or 'bpa' or 'alpha_shapes'.")
    return mesh

def add_random_transformation(pcd):
    angle = np.random.uniform(-np.pi / 10, np.pi / 10)  # Limit to +/- 18 degrees
    axis = np.random.uniform(-1, 1, 3)
    axis = axis / np.linalg.norm(axis)
    R = o3d.geometry.get_rotation_matrix_from_axis_angle(axis * angle)

    t = np.random.uniform(-0.2, 0.2, 3)  # Limit translation

    transformed_pcd = copy.deepcopy(pcd)
    transformed_pcd.rotate(R, center=(0, 0, 0))
    transformed_pcd.translate(t)

    return transformed_pcd

def sample_points_from_mesh(mesh, number_of_points):
    """
    Sample points uniformly from the surface of a mesh.

    Parameters:
        mesh (o3d.geometry.TriangleMesh): The input mesh.
        number_of_points (int): Number of points to sample.

    Returns:
        sampled_pcd (o3d.geometry.PointCloud): Point cloud with sampled points.
    """
    # Uniformly sample points from the mesh surface
    sampled_pcd = mesh.sample_points_uniformly(number_of_points=number_of_points)
    return sampled_pcd

def unit_test_floor_alignment(file_path, **kwargs):
    """
    Unit test for floor alignment.
    Applies random transformations to the point cloud and tests whether the floor is correctly aligned.

    Parameters:
        file_path (str): The path to the point cloud file.
        kwargs (dict): {depth, alpha}
    """
    print("Starting unit test for floor alignment...")
    original_pcd = load_point_cloud(file_path)
    if original_pcd is None:
        return False

    # Apply random transformations
    transformed_pcd = add_random_transformation(original_pcd)
    print("Random transformations applied.")

    # Process the transformed point cloud
    # Downsample
    pcd_down = downsample_point_cloud(transformed_pcd)
    # Remove outliers
    pcd_clean = remove_statistical_outliers(pcd_down)
    # Estimate normals
    estimate_normals(pcd_clean)
    # Detect floor plane
    plane_model, inliers = detect_floor_plane(pcd_clean)
    if plane_model is None:
        print("No floor plane detected in unit test.")
        return False

    # Align the floor to YZ plane
    aligned_pcd = align_floor_to_xz(pcd_clean, plane_model)

    # Re-detect the floor plane in the aligned point cloud
    plane_model_aligned, inliers_aligned = detect_floor_plane(aligned_pcd)
    if plane_model_aligned is None:
        print("No floor plane detected after alignment in unit test.")
        return False

    [a, b, c, d] = plane_model_aligned
    if np.allclose(a, 0, atol=1e-1) and np.allclose(b, 1, atol=1e-1) and np.allclose(c, 0, atol=1e-1):
        print("** Floor plane aligned correctly in unit test.")
    else:
        print("-- Floor plane is not aligned correctly in unit test.")
        return False


    # Check if the plane passes through y=0
    d = plane_model_aligned[3]
    if np.allclose(abs(d), 0, atol=1e-1):
        print("** Floor plane passes through y=0 in unit test.")
    else:
        print("-- Floor plane does not pass through y=0 in unit test.")
        return False

    # Also, check if the centroid of the floor plane is close to the origin
    floor_points = aligned_pcd.select_by_index(inliers_aligned)
    centroid = np.mean(np.asarray(floor_points.points), axis=0)
    print(f"Centroid: ({centroid[0]:.1f}, {centroid[1]:.1f}, {centroid[2]:.1f})")
    if np.allclose(centroid, np.zeros_like(centroid), atol=1e-1):
        print("** Floor plane is centered at the origin in unit test.")
    else:
        print("-- Floor plane is not centered at the origin in unit test.")
        return False

    print("Unit test passed: The floor plane was correctly aligned.")
    return True

def main(file_path, **kwargs):
    
    # Load the point cloud
    pcd = load_point_cloud(file_path)
    if pcd is None:
        return

    # Downsample the point cloud
    pcd = downsample_point_cloud(pcd)
    print("Point cloud downsampled.")
    print(f"Number of points after downsampling: {len(pcd.points)}")

    # Remove statistical outliers
    pcd = remove_statistical_outliers(pcd)
    print("Statistical outliers removed.")
    print(f"Number of points after outlier removal: {len(pcd.points)}")

    # Estimate normals
    estimate_normals(pcd)
    print("Normals estimated.")

    # Check if normals were estimated
    if not pcd.has_normals():
        print("Normals were not estimated after estimation step.")
        return
    else:
        print("Normals are present in the point cloud.")

    # Visualize normals
    # o3d.visualization.draw_geometries([pcd],
    #                                   window_name='Point Cloud with Normals',
    #                                   point_show_normal=True)

    # Detect the floor plane
    plane_model, inliers = detect_floor_plane(pcd)
    print("Floor plane detected.")

    # Align the floor to YZ plane and center it at the origin
    aligned_pcd = align_floor_to_xz(pcd, plane_model)
    print("Point cloud aligned.")

    # Detect the floor plane
    print("Plane equation after alignment.")
    plane_model, inliers = detect_floor_plane(aligned_pcd)

    # Save the aligned point cloud and mesh
    o3d.io.write_point_cloud('aligned_point_cloud.ply', aligned_pcd)
    print("Aligned point cloud saved.")

    # Initial surface reconstruction to get a preliminary mesh
    mesh = surface_reconstruction(aligned_pcd, method='poisson', **kwargs)
    print("Surface reconstruction completed.")
    mesh = mesh.filter_smooth_simple(number_of_iterations=kwargs["smooth_itr"])
    print("Smoothing mesh completed.")
    mesh.compute_vertex_normals()
    flipped_mesh = copy.deepcopy(mesh)
    flipped_mesh.vertex_normals = o3d.utility.Vector3dVector(-np.asarray(mesh.vertex_normals))  # Flip normals
    mesh += flipped_mesh

    # o3d.visualization.draw_geometries([initial_mesh], window_name='Initial Reconstructed Mesh', mesh_show_wireframe=True)

    # Save the final mesh
    o3d.io.write_triangle_mesh('Reconstructed_mesh.obj', mesh)
    print("Reconstructed mesh saved.")
    return mesh
```

```python
def compute_rotation_matrix_from_vectors(vec1, vec2):
    """
    Compute the rotation matrix that aligns vec1 to vec2.

    Parameters:
        vec1 (ndarray): Source vector.
        vec2 (ndarray): Destination vector.

    Returns:
        rotation_matrix (ndarray): The rotation matrix.
    """
    a, b = vec1 / np.linalg.norm(vec1), vec2 / np.linalg.norm(vec2)
    v = np.cross(a, b)
    c = np.dot(a, b)
    s = np.linalg.norm(v)

    if s == 0:
        # Vectors are parallel
        if c > 0:
            return np.eye(3)
        else:
            # Vectors are opposite
            # Find an orthogonal vector to construct the rotation matrix
            ortho = np.array([1, 0, 0]) if not np.allclose(a, [1, 0, 0]) else np.array([0, 1, 0])
            rotation_matrix = -np.eye(3) + 2 * np.outer(ortho, ortho)
            return rotation_matrix

    kmat = np.array([[0, -v[2], v[1]],
                     [v[2], 0, -v[0]],
                     [-v[1], v[0], 0]])

    rotation_matrix = np.eye(3) + kmat + np.dot(kmat, kmat) * ((1 - c) / (s ** 2))
    return rotation_matrix
```
